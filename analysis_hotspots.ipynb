```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('data/US_Accidents_June20.csv')
data = data[data['State']=='CA']
time = pd.to_datetime(data.Start_Time)
spots = ['Amenity','Bump','Crossing','Give_Way','Junction','No_Exit','Railway','Station','Stop','Traffic_Calming','Traffic_Signal']
num_spots_count = []
for s in spots:
    plt.subplots(1,2,figsize=(12,5))
    plt.xticks(fontsize=14)
    plt.suptitle('Accident Near ' + s,fontsize=16)
    plt.subplot(1,2,1)
    data.loc[data[s] == True]['Severity'].value_counts().plot.bar(width=0.5,align='center')
    plt.xlabel('Severity',fontsize=16)
    plt.ylabel('Count of Accident',fontsize=16)
    plt.xticks(fontsize=16)
    plt.yticks(fontsize=16)
    plt.subplot(1,2,2)
    data.loc[data[s] == True]['Severity'].value_counts().plot.pie(autopct='%1.0f%%')
    num_spots_count.append(data.loc[data[s] == True]['Severity'].value_counts().sum())
```
